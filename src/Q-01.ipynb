```python
import numpy as np
from scipy.misc import derivative as deriv
import plotly.express as px
import plotly.graph_objects as go
```

```python
y_hat, y = np.mgrid[0:1:0.01, 0:1:0.01]
y_hat.shape, y.shape
```

```python
mse_loss = (y_hat - y) ** 2
fig = go.Figure(data=[go.Surface(z=mse_loss, x=y, y=y_hat)])
fig.update_traces(contours_z=dict(show=True, usecolormap=True,
                                  highlightcolor="limegreen", project_z=True))
fig.update_layout(title='MSE Loss for Classification', autosize=False,
                  width=1080, height=1080,
                  scene=dict(
                      xaxis_title='Ground truth',
                      yaxis_title='Predicted Value',
                      zaxis_title='MSE Loss'
                  ))
```

```python
crossentropy_loss = -(y * np.log(y_hat) + (1 - y) * np.log(1 - y_hat))
fig = go.Figure(data=[go.Surface(z=crossentropy_loss, x=y, y=y_hat)])
fig.update_traces(contours_z=dict(show=True, usecolormap=True,
                                  highlightcolor="limegreen", project_z=True))
fig.update_layout(title='Cross-Entropy Loss for Classification', autosize=False,
                  width=1080, height=1080,
                  scene=dict(
                      xaxis_title='Ground truth',
                      yaxis_title='Predicted Value',
                      zaxis_title='Cross-Entropy Loss'
                  ))
```
